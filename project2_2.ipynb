```python
import pandas as pd
import matplotlib.pyplot as plt


df1 = pd.read_csv('cleaned_NYM_batting.csv')


top_10_players = df1.nlargest(10, 'Home_Runs')


def highlight_top_player_text(data, top_player_index, ax):
    for i, bar in enumerate(ax.patches):
        if i == top_player_index:  
            bar.set_edgecolor('black')
            bar.set_linewidth(2)
            ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.1f}', 
                    ha='center', va='bottom', fontsize=12, fontweight='bold', color='darkred')  
        else:
            ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.1f}', 
                    ha='center', va='bottom', fontsize=10, color='black')
```

Data Cleanup

The dataset was loaded from the file `cleaned_NYM_batting.csv`. The first step involved inspecting the dataset for any missing or incomplete values. The data was cleaned by:
- Removing or filling any missing values (if any).
- Ensuring all columns are of the correct data type (e.g., numeric columns like `Home_Runs`, `Batting_Average`, etc.).
- Removing any rows that could potentially introduce errors in the analysis, such as duplicate entries.

The dataset is now ready for analysis, focusing on the top 10 players in different performance metrics.


```python

plt.figure(figsize=(10, 6))  
top_10_home_runs = top_10_players['Home_Runs']
top_player_index = top_10_home_runs.idxmax()  
top_player_name = top_10_players.loc[top_player_index, 'Name']  
print(f"Top Player for Home Runs: {top_player_name}")  
ax = plt.bar(top_10_players['Name'], top_10_home_runs, color='skyblue')
plt.title('Top 10 Players by Home Runs')
plt.xlabel('Player Name')
plt.ylabel('Home Runs')
plt.xticks(rotation=45, ha='right')
highlight_top_player_text(top_10_home_runs, top_player_index, plt.gca())  
plt.tight_layout()
plt.show()
```

Graph 1: Top 10 Players by Home Runs

The first graph visualizes the top 10 players based on the number of home runs. The top player in this category is highlighted with a bold outline, and the number of home runs is displayed in red. The graph clearly shows how the top player stands out in comparison to the others.


```python

plt.figure(figsize=(10, 6))  
top_10_batting_avg = top_10_players['Batting_Average']
top_player_index = top_10_batting_avg.idxmax() 
top_player_name = top_10_players.loc[top_player_index, 'Name'] 
print(f"Top Player for Batting Average: {top_player_name}")  
ax = plt.bar(top_10_players['Name'], top_10_batting_avg, color='lightcoral')
plt.title('Top 10 Players by Batting Average')
plt.xlabel('Player Name')
plt.ylabel('Batting Average')
plt.xticks(rotation=45, ha='right')
highlight_top_player_text(top_10_batting_avg, top_player_index, plt.gca()) 
plt.tight_layout()
plt.show()
```
