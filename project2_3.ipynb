Baseball Pitchers Analysis

In this notebook, I will analyze the performance of baseball pitchers using the dataset "cleaned_baseball_pitcher.csv". This dataset contains information on various pitcher statistics, such as Wins, Earned Run Average (ERA), Strikeouts, and more.

The analysis will focus on visualizing the top 10 pitchers in three key categories: 
- Wins
- Earned Run Average (ERA)
- Strikeouts

I will also make the visualizations more informative by adding data labels and customizing the graphs.


```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('cleaned_baseball_pitcher.csv')
```

Top 10 Pitchers by Wins

In this graph, we will visualize the top 10 pitchers by the number of wins. A higher number of wins typically reflects a successful career as a pitcher.


```python

top_players_by_wins = df[['Player name', 'Win']].sort_values('Win', ascending=False).head(10)

plt.figure(figsize=(10, 6))
bars = plt.barh(top_players_by_wins['Player name'], top_players_by_wins['Win'], color='skyblue')
plt.xlabel('Number of Wins', fontsize=12)
plt.title('Top 10 Pitchers by Wins', fontsize=14)
plt.gca().invert_yaxis()  
plt.grid(axis='x', linestyle='--', alpha=0.7)  


for bar in bars:
    plt.text(bar.get_width() + 5, bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.0f}', va='center', ha='left', fontsize=11)

plt.show()
```

Top 10 Pitchers by Strikeouts

Strikeouts are a key indicator of a pitcher's dominance. In this graph, we will examine the top 10 pitchers based on the number of strikeouts they have recorded.


```python

top_players_by_strikeouts = df[['Player name', 'Strikeouts']].sort_values('Strikeouts', ascending=False).head(10)

plt.figure(figsize=(10, 6))
bars = plt.barh(top_players_by_strikeouts['Player name'], top_players_by_strikeouts['Strikeouts'], color='lightgreen')
plt.xlabel('Number of Strikeouts', fontsize=12)
plt.title('Top 10 Pitchers by Strikeouts', fontsize=14)
plt.gca().invert_yaxis()  
plt.grid(axis='x', linestyle='--', alpha=0.7)


for bar in bars:
    plt.text(bar.get_width() + 100, bar.get_y() + bar.get_height() / 2, f'{bar.get_width():,.0f}', va='center', ha='left', fontsize=11)

plt.show()
```

Conclusion

In this analysis, we visualized the performance of the top 10 pitchers based on key statistics like Wins, Earned Run Average (ERA), and Strikeouts. By using simple bar charts, we were able to effectively compare the leading pitchers across these metrics.

Overall, pitchers with high strikeouts tend to have a significant impact on the game, while low ERA pitchers showcase their ability to control the game effectively.

The next steps in this analysis could involve examining other pitching statistics or analyzing performance over time.
